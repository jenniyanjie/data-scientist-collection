### DataScienceToolBox
[datascience toolbox installation guide](http://datasciencetoolbox.org/)   
DataScience toolbox is a good virtual environment based on Ubuntu Linux that is specifically suited for doing data science. If you are using windows system, DataScienceToolBox give you a convenient environment to develop/test code in unix system.

#### step 1: install virtualbox
#### step 2: install vagrant
open windows command line prompt (powershell)
```shell
PS C:\Users\User> cd D:
PS D:\> mkdir MyDataScienceToolBox
PS D:\> cd MyDataScienceToolBox
PS D:\MyDataScienceToolBox> vagrant init data-science-toolbox/dst
PS D:\MyDataScienceToolBox> vagrant up --provider virtualbox
```   
###### possible error and solutions:
1. VT-x is disabled in the BIOS for all cpu modes:
solutions: restart PC and access to BIOS, CPU settings, enable "Inter(R) Vitualization Technology"   
how to access BIOS of windows 10:   
Start -> Setting -> Update & securities -> Recovery -> Advanced Startup -> Restart now    
Troubleshoot -> Advanced options -> Startup settings -> Restart   

#### step 3: set-up putty and login
1. install [putty.exe](https://www.chiark.greenend.org.uk/~sgtatham/putty/latest.html)   
2. open putty.exe:    
    Host Name (or IP address): 127.0.0.1    
    Port: 2222   
    Connection type: SSH   
    save the session as "Data-Science-Toolbox"   
    Click the "Open" button to access the toolbox, enter "vagrant" for both the username and the password if it asks      
    
#### step 4: configure 
the DataScienceToolBox has a shared folder with windows: D:\MyDataScienceToolBox, in DataScienceToolBox its location is: /vagrant. I would like to create two folders: **Projects** and **Data**. In the future, all my projects are put to *Projects*, data is put to *Data*.    
Then I would create soft link within DataScientist ToolBox for easy of access folder *Projects* and *Data*.
```shell
$cd /vagrant
$mkdir Projects
$mkdir Data
$ln -sfn /vagrant/Projects/ ~/
$ln -sfn /vagrant/Data/ ~/
```   

### setup unix server (DataScienceToolBox)
to install necessary python libraries in the unix environment: 
```shell
#!/bin/bash -xe
HOME="/home/ubuntu"
sudo apt-get update
sudo apt-get upgrade
sudo apt-get install python-dev python-setuptools libatlas-dev g++
#### Install requirements for Python script
# sudo python2.7 -m pip install referer_parser
### aerospike API
sudo pip install -U aerospike
### dataframe, machine learning of python
sudo pip install -U numpy
sudo apt-get install libatlas-base-dev gfortran
sudo pip install -U scipy
sudo pip install -U pandas
sudo pip install -U statsmodels
sudo pip install -U scikit-learn
### aws
sudo pip install -U boto3
### mySQL
sudo apt-get install libmysqlclient-dev
sudo pip install mysqlclient

#### Install xgboost
#1. build the shared library
sudo yum -y install git-core
git clone --recursive https://github.com/dmlc/xgboost $HOME/xgboost
cd $HOME/xgboost; make -j4
#2. build the python package for all user
# cd /home/hadoop/xgboost/python-package; python setup.py develop --user
sudo yum install python-setuptools
cd $HOME/xgboost/python-package; sudo python setup.py install
echo "PYTHONPATH=/home/ubuntu/xgboost/python-package" > $HOME/.bashrc
source $HOME/.bashrc
```


### Spark, Scala and Zeppelin (installed in DataScienceToolBox)
overall reference: [spark installation guide](https://www.tutorialspoint.com/apache_spark/apache_spark_installation.htm)
##### step 1: download scala and spark
[scala download](http://www.scala-lang.org/download/)   
[spark download](https://spark.apache.org/downloads.html)
[zeppelin download](http://zeppelin.apache.org/docs/snapshot/install/install.html)
move the download file to local folder and untar

```shell
$tar -xvf ./spark-XXX.tgz
$tar -xvf ./scala-XXX.tgz
$tar -xvf zeppelin.XXX.tgz
$sudo mv /home/ubuntu/downloads/scala-2.11.6 /usr/local/scala 
$sudo mv /home/ubuntu/downloads/spark-1.3.1-bin-hadoop2.6 /usr/local/spark
$sudo mv zeppelin.XXXX /usr/local/zeppelin
```  
##### step 2: configure the environment  
```shell
$sudo vi /etc/environment
```  

insert following (press i): 
```shell
SCALA_HOME="usr/local/scala"
SPARK_HOME="/usr/local/spark"
ZEPPELIN_HOME="usr/local/zeppelin"
```  
save and exit (press Esc, and type :wq)

```shell
$source /etc/environment
```

add the path

```shell
$vi ~/.bash_profile
```

insert following: 
```shell
if [ -d "$SCALA_HOME/bin" ] ; then
    PATH="$SCALA_HOME/bin:$PATH"
fi
 
if [ -d "$SPARK_HOME/bin" ] ; then
    PATH="$SPARK_HOME/bin:$PATH"
fi

if [ -d "$ZEPPELIN_HOME/bin" ] ; then
    PATH="$ZEPPELIN_HOME/bin:$PATH"
fi
```

save and exit (press Esc, and type :wq)

```shell
$source~/.bash_profile
```

to use spark: 
```shell
$park-shell
$pypark
$spark-submit XXX
```

to start zeppelin:
```shell
$zeppelin-daemon.sh start
``` 
by default the zeppelin server is host on http://localhost:8080. However please note: the zeppelin is installed in the virtual unix system. In order to access the hosted page from windows web browser, we have to setup port forwarding:    
open the **Vagrantfile**, add the following text somewhere around line 30:   
```python
config.vm.network "forwarded_port", guest: 8080, host: 7080 # zeppelin
```   
by doing this, we can open the zeppelin server on http://localhost:7080
to stop zeppelin:
```shell
$zeppelin-daemon.sh stop
```
